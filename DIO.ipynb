```python
# Upload dos dados para o Google Colab
df = pd.read_csv('https://raw.githubusercontent.com/Francisco-Dougllas/Curso_DataScience_Dio.me/main/conjuntos%20de%20dados/AdventureWorks.csv')
```

```python
# Informacoes 
df.info()
```

```python
# Convertendo os tipos de dados 
df['Data Venda'] = pd.to_datetime(df['Data Venda'])
df['Data Envio'] = pd.to_datetime(df['Data Envio'])
```

```python
# Qual a Receita total
print('Tivemos um total de: ', df['Valor Venda'].sum(), 'em receita')
```

```python
# Criando a coluna Custo total
df['Custo'] = df['Custo Unitário'].mul(df['Quantidade'])
df['Custo']
```

```python
# Calculo do Custo total
print(f'Tivemos um custo total de: {df["Custo"].sum().round(3)}R$')
```

```python
# Calculo do Lucro total, primeiro criaremos a coluna lucro total e epois realizamos o calculo
df['lucro'] = df['Valor Venda'] - df['Custo']
# Total do Lucro
print(f'Tivemos um lucro de: {df["lucro"].sum().round(3)} R$')
```

```python
# Criando uma coluna com total de dias para envio do produto
df['Tempo_envio'] = df['Data Envio'] - df['Data Venda']
df.head()
```

```python
# Media de envio para cada Marca
df['Tempo_envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
df.head()
```

```python
# Media do tempo de Envio
df.groupby('Marca')['Tempo_envio'].mean()
```

```python
# Analise dos Valores NaN 
df.isnull().sum()
```

```python
# Lucro por ano e por marca
df.groupby([df['Data Venda'].dt.year, 'Marca'])['lucro'].sum()
```

```python
pd.options.display.float_format='{:20,.2f}'.format
```

```python
# Reset do Index
lucro_ano = df.groupby([df['Data Venda'].dt.year, 'Marca'])['lucro'].sum().reset_index()
lucro_ano
```

```python
# Total de Produtos Vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
# Grafico total de produtos vendidos
grafico = df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
grafico.plot.barh(title='Total de Produtos Vendidos')
```

```python
# Agrupando por ano
g_ano = df.groupby(df['Data Venda'].dt.year)['lucro'].sum()
g_ano.plot.bar(title='Lucro X Ano');
```

```python
# Analise dos valore por ano
df.groupby(df['Data Venda'].dt.year)['lucro'].sum()
```

```python
# Selecionando as vendas de 2009
vendas_2009 = df[df['Data Venda'].dt.year == 2009]
vendas_2009.head()
```

```python
# Grafico Vendas de 2009
g_2009 = vendas_2009.groupby(vendas_2009['Data Venda'].dt.month)['lucro'].sum()
g_2009.plot(title='Lucro das Vendas Por Mes de 2009')
```

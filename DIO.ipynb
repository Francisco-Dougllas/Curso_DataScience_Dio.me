```python
# Convertendo os tipos de dados 
df['Data Venda'] = pd.to_datetime(df['Data Venda'])
df['Data Envio'] = pd.to_datetime(df['Data Envio'])
```

```python
# Qual a Receita total
print('Tivemos um total de: ', df['Valor Venda'].sum(), 'em receita')
```

```python
# Criando a coluna Custo total
df['Custo'] = df['Custo Unitário'].mul(df['Quantidade'])
df['Custo']
```

```python
# Calculo do Custo total
print(f'Tivemos um custo total de: {df["Custo"].sum().round(3)}R$')
```

```python
# Calculo do Lucro total, primeiro criaremos a coluna lucro total e epois realizamos o calculo
df['lucro'] = df['Valor Venda'] - df['Custo']
# Total do Lucro
print(f'Tivemos um lucro de: {df["lucro"].sum().round(3)} R$')
```

```python
# Criando uma coluna com total de dias para envio do produto
df['Tempo_envio'] = df['Data Envio'] - df['Data Venda']
df.head()
```

```python
# Media de envio para cada Marca
df['Tempo_envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
df.head()
```

```python
# Media do tempo de Envio
df.groupby('Marca')['Tempo_envio'].mean()
```

```python
# Selecionando apenas as colunas Data Venda e Lucro
dados = df[['Data Venda','lucro']].groupby('Data Venda').sum()
dados.reset_index(inplace=True)
dados
```

```python
# Iremos Deleta a nossa data de venda e depois iremos criar uma nova data ficticia para Series Temporais
dados.drop("Data Venda", axis=1, inplace=True)
```

```python
# Criando um novo indice data apenas com dias uteis
# Para os dados ficarem um pouco relacionados com a data iremos utiliza a data de 2008-02-20 a 2009-12-25
indice = pd.date_range('2008-02-20', periods=len(dados), freq='B')
indice
```

```python
# Criando uma nova serie
serie = pd.Series(dados['lucro'].values, index=indice )
serie[:10]
```

```python
# Tipo de serie
type(serie)
```

```python
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 10,4
```

```python
# Plot exemplo
serie.plot(figsize=(10, 5));
```

```python
# Criando uma media movel para comaparacao
media_movel = serie.rolling(window=5)
media_movel = media_movel.mean()
plt.figure(figsize=(12, 5))
plt.plot(media_movel);
```

```python
# Comparando a Serie com A Media
plt.figure(figsize=(12, 5))
plt.plot(serie[-100:], label='Série Real')
plt.plot(media_movel[-100:] ,color='red', label='Média Móvel 5 Dias')
plt.legend(loc='best');
```

Analise e decomposicao dos nossos dados

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decomposicao = seasonal_decompose(serie, period=7)
```

```python
# Plotagem com legendas dos graficos
plt.figure(figsize=(12,6))
plt.subplot(411)
plt.plot(serie, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(decomposicao.trend, label='Tendência')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(decomposicao.seasonal,label='Sazonalidade')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(decomposicao.resid, label='Resíduos')
plt.legend(loc='best')
plt.tight_layout();
```
